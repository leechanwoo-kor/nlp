# Word2Vec 임베딩 실습

```python
## gensim 라이브러리 사용
from gensim.models import word2vec
```

```python
## corpus 저장소 지정
corpus_fname = "/content/drive/MyDrive/Word2vec/Corpus/ratings_mecab.txt"
```

```python
## model 저장소 지정
model_fname = "/content/drive/MyDrive/Word2vec/Model/word2vec"
```

```python
## 한 단어씩 listup
%time corpus = [sent.strip().split(" ") for sent in open(corpus_fname, 'r').readlines()]
```

```python
## corpus 길이 확인
len(corpus)
```

```python
## 모델 생성 (10차원, CPU=4개, Skip-gram)
model = word2vec.Word2Vec(corpus,
                          size=10,
                          workers=4,
                          sg=1)
```

```python
## 가장 유사한 상위 단어 확인
model.most_similar("희망", topn=5)
```

```python
model.most_similar("절망", topn=5)
```

```python
model.most_similar("학교", topn=5)
```

```python
## 단어 사전에서 상위 10개만 보기
vocab = model.wv.vocab
sorted(vocab, key=vocab.get, reverse=True)[:10]
```

```python
## 단어 벡터값 확인
model.wv['영화']
```

```python
## 모델 저장
model.save(model_fname + '_size_10')
```
